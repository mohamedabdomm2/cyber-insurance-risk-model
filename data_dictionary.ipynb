# Data Dictionary

<div class="alert alert-info" role="alert">
  <p>
    <center><b>📇 DATASET CARD: Cyber Insurance Risk Modeling</b></center>
  </p>

  <p>
    This dataset contains 100,000 synthetic, highly realistic business records designed to train an underwriting model for a Cyber Insurance provider. The data maps a company's firmographics, cybersecurity defensive posture, and data exposure to their historical likelihood of suffering a catastrophic cyber breach and filing an insurance claim.
  </p>

  <hr>

  <p><b>Part 1: The Identity Layer (Firmographics)</b><br>
  <i>These features define the "size and shape" of the target.</i></p>
  <ul>
    <li><b>company_id</b> (Int): Primary Key. Unique identifier for the client.</li>
    <li><b>industry_sector</b> (Categorical): The company's vertical (e.g., Healthcare, Finance, Retail, Technology). <i>Note: Healthcare & Finance are historically targeted more frequently.</i></li>
    <li><b>employee_count</b> (Int): Total number of staff (represents the number of "phish-able" targets).</li>
    <li><b>annual_revenue</b> (Float): Total yearly revenue in USD. Used to calculate potential Business Interruption losses.</li>
    <li><b>operating_budget</b> (Float): Yearly operating expenditure.</li>
    <li><b>years_in_operation</b> (Int): Company maturity. Older companies may have more insecure "legacy" systems.</li>
    <li><b>hq_location</b> (Categorical): Country of headquarters, which dictates data privacy law jurisdiction (e.g., GDPR, CCPA).</li>
  </ul>

  <p><b>Part 2: The Defensive Layer (Security Posture)</b><br>
  <i>These are the Risk Mitigators. Stronger defenses should correlate negatively with claim probability.</i></p>
  <ul>
    <li><b>mfa_status</b> (Binary: 0/1): Is Multi-Factor Authentication enforced across all internal systems?</li>
    <li><b>edr_deployed</b> (Binary: 0/1): Does the company use Endpoint Detection & Response (EDR) software?</li>
    <li><b>patch_velocity</b> (Int): Average number of days it takes the IT team to patch a "Critical" vulnerability.</li>
    <li><b>open_vulnerabilities</b> (Int): Current number of unpatched flaws.</li>
    <li><b>backup_type</b> (Categorical): How data is backed up (Offline/Air-gapped, Cloud-only, On-premise).</li>
    <li><b>training_freq</b> (Int): How often employees get security awareness training (0 = Never, 4 = Quarterly).</li>
    <li><b>pen_test_age</b> (Int): Months since the last professional 3rd-party "Hacker" test was conducted.</li>
  </ul>

  <p><b>Part 3: The Crown Jewels Layer (Data Exposure)</b><br>
  <i>These features define the severity and attractiveness of the target to hackers.</i></p>
  <ul>
    <li><b>pii_count</b> (Int): Total count of Personally Identifiable Information records stored by the company.</li>
    <li><b>database_storage_tb</b> (Float): Total data footprint in Terabytes.</li>
    <li><b>pci_data_flag</b> (Binary: 0/1): Does the company process Credit Card information?</li>
    <li><b>phi_data_flag</b> (Binary: 0/1): Does the company process Protected Health Information?</li>
    <li><b>cloud_adoption_pct</b> (Float): Percentage of infrastructure hosted in the Cloud (AWS/Azure) vs. On-premise.</li>
    <li><b>subdomain_count</b> (Int): Number of public-facing web addresses (defines the external "Attack Surface").</li>
  </ul>

  <p><b>Part 4: The Ground Truth (Incident History & The Target)</b><br>
  <i>This data represents the aftermath. For non-breached companies (y=0), these fields (except the target) are NaN.</i></p>
  <ul>
    <li><span style="color: red">🎯</span> <b>is_claim_filed</b> (Binary: 0/1): <b>TARGET VARIABLE.</b> Did the client suffer a breach and use their insurance?</li>
    <li><b>breach_type</b> (Categorical): The vector of attack (Ransomware, Phishing, DDoS, Insider Threat).</li>
    <li><b>total_loss_val</b> (Float): The total financial cost of the claim.</li>
    <li><b>records_stolen</b> (Float): Actual number of records leaked in the event.</li>
    <li><span style="color: red">🚨</span> <b>forensics_retained</b> (Float: 1.0 or NaN): Did the company hire a digital forensics team post-breach?</li>
    <li><span style="color: red">🚨</span> <b>post_breach_pr_cost</b> (Float): Money spent on Public Relations to save the brand image after a hack.</li>
  </ul>
</div>
